# 3.Check whether all the zones generate the same amount of sales

H0 : zones generate the same amount of sales

H1 : zones generate dfferent amount of sales

```python
f_val,p_val=stats.f_oneway(df["Zone - A"], df["Zone - B"], df["Zone - C"],df["Zone - D"])
print("F value:",f_val)
print("P value:",p_val)
```

```python
if(p_val<0.05):
    print("Reject H0 differnt zones generate dfferent amount of sales")
else:
    print("Accept H0 all zones generate same amount of sales")
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column=["Zone - A","Zone - B","Zone - C","Zone - D"])
plt.show()
```
